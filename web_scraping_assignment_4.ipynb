```python
driver.get('http://www.bcci.tv/')
time=10
```

```python
int_click=driver.find_element_by_xpath('/html/body/nav/div[1]/div[2]/ul[1]/li[2]/a')
driver.get(int_click.get_attribute('href'))
time=5
```

```python
title1=[]
series1=[]
place1=[]
date1=[]
time1=[]
```

```python
for i in range(2):
    title_tags1=driver.find_elements_by_xpath("//span[@class='matchOrderText ng-binding ng-scope']")
    for j in title_tags1:
        title1.append(j.text.split('-')[0])
    series_tags1=driver.find_elements_by_xpath("//span[@class='ng-binding']")
    for k in series_tags1:
        series1.append(k.text)
    place_tags1=driver.find_elements_by_xpath("//span[@class='ng-binding ng-scope']")
    for l in place_tags1:
        place1.append(l.text)
    date_tags1=driver.find_elements_by_xpath("//h5[@class='ng-binding']")
    for m in date_tags1:
        date1.append(m.text)
    time_tags1=driver.find_elements_by_xpath("//h5[@class='text-right ng-binding']")
    for n in time_tags1:
        time1.append(n.text)
    if i<1:
        try:
            more_tags1=driver.find_element_by_xpath("//button[@class='match-btn btn-red d-flex align-items-center justify-content-center mx-auto mt-3']")
            more_tags1.click()
        except NoSuchElementException:
            delay=15
            WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME,'match-btn btn-red d-flex align-items-center justify-content-center mx-auto mt-3')))
            more_tags1=driver.find_element_by_xpath("//button[@class='match-btn btn-red d-flex align-items-center justify-content-center mx-auto mt-3']")
            more_tags1.click()
```
